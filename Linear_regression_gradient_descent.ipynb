```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
x = data.iloc[:,0]
y = data.iloc[:,1]
```

```python
plt.scatter(x,y)
```

```python
m = 0
c = 0
epochs = 1000
alpha = 0.0001
n = float(len(x))
```

```python
for i in range(epochs):
    y_pred = m*x + c
    D_m = (-2/n) * sum(x * (y - y_pred))
    D_c = (-2/n) * sum(y - y_pred)
    m = m - alpha*D_m
    c = c - alpha*D_c
    
print(m,c)
```

```python
y_pred = m*x + c
plt.scatter(x, y) 
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='red')
plt.show()
```
